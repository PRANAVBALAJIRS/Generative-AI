## ChatCompletion API and Completion API

```python
prompt = "Hello, How are you?"

response = openai.ChatCompletion.create(
    model = 'gpt-3.5-turbo',
    messages = [
        {"role": "system", "content": "You are a helpful assistant"},
        {"role": "user", "content": prompt}
   ]
)

print(response)
```

```python
print(response["choices"][0]["message"]["content"])
```

```python
prompt_1 = "Hello, How are you?. My name is Pranav Balaji and I am AI and ML undergrad"
prompt_2 = "Tell me about me and what do you think about me"
response = openai.ChatCompletion.create(
    model = 'gpt-3.5-turbo',
    messages = [
        {"role": "system", "content": "You are a helpful assistant"},
        {"role": "user", "content": prompt_1},
        {"role": "user", "content": prompt_2}
   ]
)

print(response)
```

```python
prompt = "Tell me about your founders"

response = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = [
        {"role": "system", "content": "You are a helpful assistant"},
        {"role": "user", "content": prompt}
    ],
    max_tokens = 25, #Max no of output tokens
    temperature = 0.6, #Randomness in response
    n = 2 #No of responses
)

print(response)
```

```python
functions = [
    {
        "name": "get_current_weather",
        "description": "Get the current weather in a given location",
        "parameters": {
            "type": "object",
            "properties": {
                "location": {
                    "type": "string",
                    "description": "The city and state, e.g. San Francisco, CA"
                },

            },
            "required": ["location"],
        },
    }
]
```

```python
user_message = "Hi there"
messages = []
messages.append({"role": "user", "content": user_message})

completion = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = messages
)
```

```python
print(completion.choices[0].message)
```

```python
user_message = "What is the temperature in Chennai"

messages.append({"role": "user", "content": user_message})

completion = openai.ChatCompletion.create(
    model = "gpt-3.5-turbo",
    messages = messages,
    functions = functions
)
```

```python
response = completion.choices[0].message
```

```python
function_name = response['function_call']['name']
print(function_name)
```
